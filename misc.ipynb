```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from scipy import special
```

```python
theta = np.linspace(0, 1, 101)
```

```python
k = 0.5
plt.plot(k * stats.beta(a=16, b=4).pdf(theta) + (1-k) * stats.beta(a=10, b=10).pdf(theta), label='prior')


y = 79
n = 120

a1 = 16 + y
a2 = 10 + y
b1 = 4 + (n - y)
b2 = 10 + (n - y)
A = k * special.beta(a1, b1) / special.beta(16, 4)
B = (1 - k) * special.beta(a2, b2) / special.beta(10, 10)
q = A / (A+B)
print a1, a2, b1, b2, q
print a1 /float(a1+b1), a2 / float(a2+b2)
plt.plot(stats.beta(a=a1, b=b1).pdf(theta), label='posterior_1', ls='--')
plt.plot(stats.beta(a=a2, b=b2).pdf(theta), label='posterior_2', ls='--')

plt.plot(q * stats.beta(a=a1, b=b1).pdf(theta) + (1-q) * stats.beta(a=a2, b=b2).pdf(theta), label='posterior')

plt.legend()
```
